```python
import requests
import os

openai_api_key = "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"
if openai_api_key is None:
    raise ValueError("OpenAI API key is not set in environment variables.")

url = "https://api.openai.com/v1/chat/completions"

headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {openai_api_key}"
}

def categorize_app(app_name, app_description):
    data = {
        "model": "gpt-3.5-turbo",
        "messages": [
            {"role": "system", "content": "You are a model categorizing apps."},
            {"role": "user", "content": f"Categorize the app {app_name}. Description: {app_description}"}
        ]
    }

    response = requests.post(url, headers=headers, json=data)

    # Check if the request was successful
    if response.status_code == 200:
        generated_text = response.json()['choices'][0]['message']['content'].strip()
        # Assume the category is the last word in the generated text
        predicted_category = generated_text.split()[-1]
        return predicted_category
    else:
        print("Error:", response.status_code, response.text)
        return None

# Your app names and descriptions go here
your_apps_data = {
    'Delta': 'Delta helps with booking flights.',
    'TripAdvisor': 'TripAdvisor is for exploring new places.',
    'Avis': 'Avis is a car rental service.'
    # Add more apps as needed
}

# Get the predicted categories
predicted_categories = {app_name: categorize_app(app_name, app_description) for app_name, app_description in your_apps_data.items()}

# Display the results
for app_name, predicted_category in predicted_categories.items():
    print(f"{app_name} is categorized under: {predicted_category}")
```
